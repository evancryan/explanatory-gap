```python
import pandas as pd
import plotly.express as px
```

```python
df_scores = pd.read_csv("scores.csv")
print(df_scores.head())

min_conf = df_scores["Confidence"].min()
max_conf = df_scores["Confidence"].max()

fig = px.scatter(
    df_scores,
    x="Focus",
    y="Magnitude",
    color="Confidence",
    color_continuous_scale=["black", "red"],
    range_color=[min_conf, max_conf],
    title="Interactive Plot of 3 Scores",
    hover_name=df_scores["Title"].apply(lambda x: str(x)[:50] + "..." if len(str(x)) > 50 else str(x)),
    hover_data=["Year"]
)

fig.add_hline(y=0, line_dash="dash", line_color="black", opacity=0.25)
fig.add_vline(x=0, line_dash="dash", line_color="black", opacity=0.25)

fig.update_layout(
    xaxis_title="<-- Subjective / Objective -->",
    yaxis_title="<-- Micro / Macro -->",
    xaxis_range=[-1, 1],
    yaxis_range=[-1, 1],
    width=650,
    height=550,
    margin=dict(l=50,r=50,t=50,b=50),
    yaxis_scaleanchor="x"
)

fig.show()
```
